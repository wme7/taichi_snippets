# Interact with Python libraries, such as Numpy
Ref: Yuanming Hu, Taichi Cookbook 001, 27.07.22

```python
import taichi as ti
from matplotlib import pyplot as plt
```

```python
ti.init(arch=ti.cpu)
```

```python
n = 8192
factors = ti.field(dtype=ti.i32, shape=n)
```

```python
@ti.kernel
def number_of_factors():
    for i in range(1, n): 
        counter = 0
        for j in range(1, int(ti.floor(ti.sqrt(i))) + 1):
            if i % j == 0:
                if j * j == i:
                    counter += 1
                else:
                    counter += 2
        factors[i] = counter
```

```python
number_of_factors()
```

```python
plt.plot(factors.to_numpy(), '.')
plt.show()
```
